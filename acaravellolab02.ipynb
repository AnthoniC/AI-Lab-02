```python
import random
random.seed()

def restart():
    state = [0]*8
    for i in range(8):
        r = random.randrange(8)
        state[i] = r
    return state
    
state = restart()
print(state)
```

```python
import math
def isAttacking(s, q1, q2):
    hdelta = math.fabs(s[q1]-s[q2])
    if hdelta == 0.0:  
        return True
    vdelta = math.fabs(q1-q2)
    
    if hdelta == vdelta: 
        return True
    return False
```

```python
def objf(state):
    sum = 0
    for q1 in range(8):
        for q2 in range(8):
            if q1==q2:
                continue
            if not isAttacking(state,q1,q2):
                sum += 1
    return sum
```

```python
def display(board):
    for row in range(8):
        rowStr = str(board[row])
        for col in range (8):
            if board[row] == col: 
                rowStr += ' Q'
            else:
                rowStr += ' _'
        print(rowStr)
    print(str(objf(board))+"\n")

display(state)
```

```python
def nextStates(state):
    lst = []
    for r in range(8):
        skpi = state[r]
        other = list(range(8))
        other.remove(skpi)
        for o in other:
            nstate = list(state)
            nstate[r] = o
            lst.append(nstate)
    return lst
            
n = nextStates([5,3,6,0,7,1,4,7])
for i in n:
    display(i)
```

```python
# Finds the next best k number of states and returns them as a list
def bestNextStates(states, k): 
    nextGen = []
    for i in states:
        nextGen.extend(nextStates(i))
    m = list(map(objf, nextGen))
    nex = [] # Could probably make this a set to improve
    for i in range(k):
        nex.append(nextGen[m.index(max(m))])
        nextGen.pop(m.index(max(m)))
        m.pop(m.index(max(m)))
    return nex

bestNextStates([[5,3,6,0,7,1,4,7],[5,3,6,0,7,1,7,2]], 5)
```

```python
def beamSearch(k):
    states = []
    count = 0
    goalFound = False
    for i in range(k):
        states.append(restart())
    
    while (count<100):
        states = bestNextStates(states, k)
        count += 1
        
        
        if(objf(states[0]) == 56):
            goalFound = True
            break
    return goalFound, count, objf(states[0])

beamSearch(3)
    
```

Answers to questions posed

1. Is k-beam search complete?
    No, It still suffers from the problem of hill climbing algorithms; which is that it can get stuck at a local max or min and not find the global max or min we are searching for. 

2. Is there a point where a higher k no longer provides value? 
    Yes. This is because as you increase the k value, you will need to find more next states, which will increase the time you have to spend at any given step. The best k value will be lowest value that should give you a decent distribution of the state-space; I use the term should because if the starting states are randomly chosen then we cannot guarantee an even distribution across the state-space. 

3. Is k-beam search effective when the number of states in infinite?
    It depends on how you define effective, in theory it will always find the best solution because it will find the best solution in one infinite starting states; but it would have a infinite memory usage, which means it is impossible, also it would increase the time to find the next states to infinity. At that point it would just be a random selection of states until you find the best one.
